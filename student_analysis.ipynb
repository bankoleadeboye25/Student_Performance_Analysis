```python
#HISTogram of Scores
import matplotlib.pyplot as plt
plt.hist(data["Score"], bins=10, edgecolor="black")
plt.title("Distribution of JSS1 Student Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

Histogram Interpretation
The distribution of scores shows that most students scored between 50 and 60, indicating that the majority of students fall within the average performance range. There are relatively few students at both the lower and higher ends of the score spectrum, suggesting limited variation in extreme performance levels.

```python
#Box Plot of Scores
plt.boxplot(data["Score"], vert=False)
plt.title("Box Plot of JSS1 Student Scores")
plt.xlabel("Scores")
plt.show()
```

The boxplot analysis indicates that the median score is approximately 54, which is consistent with the mean, suggesting a fairly symmetrical distribution of scores. The majority of students scored within a narrow range of approximately 50 to 58, indicating relatively low variability among average-performing students. However, a few outliers are present on both the lower and upper ends of the distribution, highlighting the existence of both underperforming and high-achieving students.
